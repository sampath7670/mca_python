```python
class Area:
    length=8
    breadth=5
    def calc(self):
        ans=self.length*self.breadth
        return ans
```

```python
class Area2:
    def __init__(self,len,bth):
        self.l=len
        self.b=bth
    def calc(self):
        ans=self.l*self.b
        return ans
```

```python
o1 = Area()
o1.length = 9
o1.breadth = 8
res = o1.calc()
print(res)
```

```python
class Dog:
    def __init__(self,breed,color,age):
        self.breed=breed
        self.color=color
        self.age=age
    def show (self):
        print("The Dog details")
        print("Breed:",self.breed)
        print("AGE:",self.age)
        print("Color:",self.color)
bruno=Dog("German Shepherd","Black",5)
buddy=Dog("ROBO","Brown",3)
sammy=Dog("Vodaphone","White",4)
alex=Dog("Rottweiler","Black and Brown",6)
bruno.show()
buddy.show()
sammy.show()
alex.show()
```

```python

doglist=[bruno,buddy,sammy,alex]
age=0
eldest_dog=None
for dog in doglist:
    if age<=dog.age:
        age=dog.age
        eldest_dog=dog
print(eldest_dog.show())
```

```python
class Pokemon:
    def __init__(self,name,type='Normal',hp=10,atk=10,dnf=10):
        self.name=name
        self.type=type
        self.hp=hp
        self.atk=atk
        self.dnf=dnf
    def stats(self):
        print("Pokemon Details")
        print(f'Name:{self.name}')
        print(f'Type:{self.type}')
        print(f'HP:{self.hp}')
        print(f'Attack:{self.atk}')
        print(f'Defense:{self.dnf}')
    def overall(self):
        return self.hp+self.atk+self.dnf
    def is_powerful(self):
        if self.overall()>100:
            return True
        return False
pikachu=Pokemon("Pikachu","Electric",35,55,40)
snorlax=Pokemon("Snorlax","Normal",160,110,65)
charizard=Pokemon("Charizard","Fire",78,84,78)
bulbasaur=Pokemon("Bulbasaur","Grass",45,49,49)
ratata=Pokemon("Rattata")
Pokemons=[pikachu,snorlax,charizard,bulbasaur,ratata]
print("\nStats")
for poke in Pokemons:
    poke.stats()
    print(f'Overall Stats: {poke.overall()}')
    print(f'Is Powerful:{poke.is_powerful()}')
    print("/////////////////////////////////")
```

```python
class Calc:
    def run(self,num1,operator,num2):
        match operator:
            case '+':return self.add(num1,num2)
            case '-':return self.sub(num1,num2)   
            case '*':return self.mul(num1,num2)
            case '/':return self.div(num1,num2)
    def add(self,a,b): return a+b
    def sub(self,a,b): return a-b
    def mul(self,a,b): return a*b       
    def div(self,a,b):
        if b==0: return "ðŸš¨ Can't Divided by 0"
        return a/b
```

```python
c=Calc()
while True:
    a=int(input("Enter first number: "))
    b=int(input("Enter second number: "))
    op=input("Enter operator (+,-,*,/): ")
    ans= c.run(a,op,b)
    print(ans)
    if input("Continue") !='y':
        break
```
